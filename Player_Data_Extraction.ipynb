<h2>Imports</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
```

<h2>NBA Player Directory</h2>

```python
letters = np.array(list('abcdefghijklmnopqrstuvwxyz'))
letters = ['a','b','c','d','e','f']
# letters = np.array(list('ghijklmnopqrstuvwxyz'))
# letters = np.array(list('jklmnopqrstuvwxyz'))
# letters = np.array(list('wxyz'))
letters
```

```python
base_player_directory_url = 'https://www.basketball-reference.com/players/'
players = []
for letter in letters:
    try:
        page_query = requests.get(base_player_directory_url+letter+'/')
        page_html = bs(page_query.content,'html.parser')
        player_table = page_html.find('table',class_='sortable').find('tbody')
        for player_a_tag in player_table.find_all('tr'):
            players.append(player_a_tag.find('a').get_text().lower().replace('.','').replace("'","").split())
    except:
#         Will print x because there is no player with last name starting with x
#         print(letter)
        continue
print(len(players))
# print(players)
```

<h2>Access Stats Pages for Each Player</h2>

```python
def to_int(float_num):
    return int(float_num)
to_int_vectorized = np.vectorize(to_int)

def to_float(num):
    return float(str(num))
to_float_vectorized = np.vectorize(to_float)
```

```python
# players_stats_df_dict = {}
for player in players:
    try:
        
    #     Getting the player stats html
        first_name = player[0]
        last_name = player[1]
        player_stats_url = base_player_directory_url + last_name[0]+'/'+last_name[:5]+first_name[:2]+'01.html'
        player_stats_df_list = pd.read_html(player_stats_url, attrs = {'id':'per_game'})
        if 'NBA' not in player_stats_df_list[0]['Lg'].values:
            continue


    #     Cleaning player age and aggregating data
        player_stat_nan_indices = np.isnan(player_stats_df_list[0]['Age'].values)
        player_stat_useless_rows = player_stats_df_list[0][player_stat_nan_indices]
        player_stats_df_list[0].drop(player_stat_useless_rows.index.values, inplace=True)
        player_stats_df_list[0]['Age'] = to_int_vectorized(player_stats_df_list[0]['Age'].values)
        try:
            player_stats_df_list[0] = player_stats_df_list[0].groupby(['Age']).mean()
        except:
            injury_rows = player_stats_df_list[0]['Tm'].str.contains('Did Not Play')
            player_stats_df_list[0].drop(player_stats_df_list[0][injury_rows].index.values,inplace=True)
            player_stats_df_list[0].drop(columns=['Tm','Lg','Season','Pos'], axis=1,inplace=True)
            player_stats_df_list[0].iloc[:,:] = to_float_vectorized(player_stats_df_list[0].iloc[:,:])
            player_stats_df_list[0] = player_stats_df_list[0].groupby(['Age']).mean()
            print('players injury seasons removed')
            
            
    # Selecting useful/manipulating statistics
        player_stats_df = player_stats_df_list[0][['G','FGA','FG%','FTA','FT%','TRB','AST','PTS']]
        player_stats_df['FT%'] = player_stats_df['FT%'] * 100
        player_stats_df['FG%'] = player_stats_df['FG%'] * 100


    #     NaN Handling
        player_stats_df = player_stats_df.interpolate().fillna(method='bfill')
    
    
    #   Storing player stats in pickle
        player_stats_df.to_pickle('player_stats/'+first_name+'_'+last_name+'.pickle')
        print(player)
    except Exception as e:
        print(player, ' had an exception: \n',e)
```
